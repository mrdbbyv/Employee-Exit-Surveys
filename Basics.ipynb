# Clean and Analyze Employee Exit Surveys

This project will be focused on cleaning and analyzing exit surveys from employees of [the Department of Education](https://en.wikipedia.org/wiki/Department_of_Education_(Queensland)), Training and Employment and the Technical and Further Education (TAFE) body of the Queensland government in Australia. The TAFE exit survey can be found [here](https://data.gov.au/dataset/ds-qld-89970a3b-182b-41ea-aea2-6f9f17b5907e/details?q=exit%20survey) and the survey for the DETE can be found [here](https://data.gov.au/dataset/ds-qld-fe96ff30-d157-4a81-851d-215f2a0fe26d/details?q=exit%20survey).

As a data analyst, our stakeholders want us to combine the results for both surveys to answer the following questions:

- Are employees who only worked for the institutes for a short period of time resigning due to some kind of dissatisfaction? What about employees who have been there longer?

- Are younger employees resigning due to some kind of dissatisfaction? What about older employees?

## Creating `institute_service` column

We need to create a column containing the length of time an employee spent at work in both dataframes. So these questions need to be answered.

Are employees who have only worked for the institutes for a short period of time resigning due to some kind of dissatisfaction? What about employees who have been at the job longer?
The `tafe_resignations` dataframe already contains a "service" column, which we renamed to `institute_service`.

Below, we calculate the years of service in the `dete_survey_updated` dataframe by subtracting the `dete_start_date` from the `cease_date` and create a new column named `institute_service`.

```python
dete_resignations['institute_service'] = dete_resignations['cease_date'] - dete_resignations['dete_start_date']
```

## Finding employees who resigned

Next, we will identify any employees who resigned because they were dissatisfied. Columns from `tafe_survey_updated` and `dafe_survey_updated` will be used to create a new `dissatisfied` column which will contain *'True'*, *'False'* and *'NaN'* values. 

```python
tafe_resignations['Contributing Factors. Dissatisfaction'].value_counts()
```

```python
tafe_resignations['Contributing Factors. Job Dissatisfaction'].value_counts()
```

```python
def update_vals(x):
    if x == '-':
        return False
    elif pd.isnull(x):
        return np.nan
    else:
        return True
tafe_resignations['dissatisfied'] = tafe_resignations[['Contributing Factors. Dissatisfaction', 'Contributing Factors. Job Dissatisfaction']].applymap(update_vals).any(axis=1, skipna=False)
tafe_resignations_up = tafe_resignations.copy()

tafe_resignations_up['dissatisfied'].value_counts(dropna=False)
```

```python
dete_resignations["dissatisfied"] = dete_resignations[['job_dissatisfaction',
       'dissatisfaction_with_the_department', 'physical_work_environment',
       'lack_of_recognition', 'lack_of_job_security', 'work_location',
       'employment_conditions', 'work_life_balance',
       'workload']].any(axis=1, skipna=False)
dete_resignations_up = dete_resignations.copy()
print(dete_resignations_up["dissatisfied"].value_counts(dropna=False))
```

## Combining two dataset

## `institue_service` column cleaning

Next, we'll clean the `institute_service` column and categorize employees according to the following definitions:

New: Less than **3** years in the workplace
Experienced: **3-6** years in the workplace
Established: **7-10** years in the workplace
Veteran: **11** or more years in the workplace

Understanding employee's needs according to career stage instead of age is more effective.

```python
combined['institute_service'] = combined['institute_service'].astype('str').str.extract(r'(\d+)')
```

```python
combined['institute_service'] = combined['institute_service'].astype('float')
```

```python
def map_each_year(val):
    if pd.isnull(val):
        return np.nan
    elif 3 <= val < 7:
        return "Experienced"
    elif 7 <= val <11:
        return "Established"
    elif val >= 11:
        return "Veteran"
    else:
        return "New"
```

```python
combined['service_cat'] = combined['institute_service'].apply(map_each_year)
```

## Some analysis

We assume the missing values in the dissatisfied column can be replaced with *False* (the most frequent one). Then, we will calculate the percentage of employees who resigned due to dissatisfaction in each `service_cat` group and plot the results.

```python
combined['dissatisfied'].value_counts(dropna=False)
```

```python
combined['dissatisfied'] = combined['dissatisfied'].fillna(value=False)
```

```python
pv_dissatisfied = combined.pivot_table(index='service_cat', values='dissatisfied')
```

```python
%matplotlib inline
pv_dissatisfied.plot(kind='bar')
```

## Conclusion

Analysis shows that employees with **7** or more years of experience are more likely to resign due to some sort of dissatisfaction than employees with less than **7** years of service.
